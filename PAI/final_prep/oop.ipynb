## class init

````python
class child:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = child("John", 36)
print(p1.name)
print(p1.age)
````

### multiple

```mermaid
graph TD
A --> C
B --> C
```

````python
class mother:
    motherName = ''
    def mother(self):
        print(self.motherName)
    
class father:
    fatherName = ''
    def father(self):
        print(self.fatherName)

class son(mother, father):
    def parents(self):
        print(self.motherName, self.fatherName)

s1  = son()
s1.fatherName = "f"
s1.motherName = "m"

s1.parents()
````

#### multi-level

```mermaid
graph LR
a-->b-->c
```


````python
class A:
    def __init__(self,name):
        self.name = name

class B(A):
    def __init__(self,name,age):
        self.age = age
        A.__init__(self,name) #invoke A via B

class C(B):
    def __init__(self,name,age,id):
        self.id = id
        B.__init__(self,name,age) #invoke B via C
    
    def printInfo(self):
        print(self.name,self.age,self.id)

s1 = C("muneeb",23,765)

s1.printInfo()
````

````python
class parent:
    def parentFunc(self):
        print("defined in parent")

class child(parent):
    def childFunc(self):
        print("defined in child")

child = child()
child.parentFunc()
````

### `Super` keyword

- allows parent method redefinition
- aka. "method overriding"


````python
class parent:
    def show(self):
        print("defined in parent")

class child(parent):
    def show(self):
        super().show()
        print("called in child")

child = child()
child.show()
````

## polymorphism

````python
def product(a,b):
    return a*b
def product(a,b,c):
    return a*b*c

## with classes

class A:
    def func(self):
        print("A")
class B:
    def func(self):
        print("B")
oA = A()
oB = B()

for x in (oA,oB):
    x.func()
````

## `*args` 

- arbitrary arguments
- `list` as arguments

````python
def my_function(*kids):
    print("The youngest child is " + kids[2])
my_function("Sameer", "Talha", "Sohail")

# list as arg
def my_function_two(kids):
    print("The youngest child is " + kids[2])
list = ["Sameer", "Talha", "Sohail"]
my_function_two(list)
````

## `*kwargs`

- keyword arguments
- unordered arguments

````python
def my_function(child3, child2, child1):
    print("The youngest child is " + child3)
my_function(child1 = "Emil", child2 = "Tobias", child3 = "Linus")
````

## enumerate
- creates a pair of tuples, where the first val is the index

````python
x = ('apple', 'banana', 'cherry')
y = enumerate(x)

for val in y:
    print(val)
````

## zip
- creates an ordered tuple of tuples, based on position

````python
a = ("John", "Charles", "Mike")
b = ("Jenny", "Christy", "Monica")
x = zip(a, b)
for val in x:
    print(val)
print("-----")
y = enumerate(zip(a, b))
for val in y:
    print(val)
````

```python
import pandas as pd
source = "https://raw.githubusercontent.com/R4F4I/semester-3-FAST-K240009/refs/heads/master/PAI/LAB/assignments/lab_6_7_8/Software%20houses%20PK%20.csv"
df = pd.read_csv(source)
```

1. Analyzing where most software houses are located in Pakistan. Are they concentrated in major cities like Karachi, Lahore, Islamabad, or are there emerging hubs in smaller cities?

```python
city_counts = df['City'].value_counts()
print("Frequency of software houses per city:\n")
print(city_counts)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
city_counts.plot(kind='bar',color='blue')
plt.title('Distribution of Software Houses Across Cities in Pakistan')
plt.xlabel('City')
plt.ylabel('Number of Software Houses')
plt.tight_layout()
plt.show()
```

2. Determining the industries these software houses cater to. Are they primarily focused on software development for finance, healthcare, e-commerce, or other sectors?

```python
import numpy as np

# remove all -
services = df['Services'].replace(' ',np.nan).dropna()
services = df['Services'].replace('-',np.nan).dropna()


# given that the services are presented as a list "a,b,c" convert it into a list ["a","b","c"]
services = services.str.split(',')

# given that it is a 2d array
# ["a","b","c"],
# ["d","e","f"],
# ["g","h","i"]
# convert it to
# ["a","b","c","d","e","f","g","h","i"]
services = services.explode()

# clean the string
services = services.str.strip().str.lower()

# group by their frequency
services_counts = services.value_counts()



print(services_counts.head(21))
```

```python
plt.figure(figsize=(12,7))
services_counts.head(21).plot(kind='bar',color='blue')
plt.title("Industries offered")
plt.xlabel('service')
plt.ylabel('Number of software house')
plt.show()
```

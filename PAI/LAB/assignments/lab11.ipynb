Load an image from file and display it, convert to grayscale, resize it to specific size using OpenCV.


```python
import cv2
img = cv2.imread('mountains.jpg')
if img is None:
    print("---------------------------------------------------------")
    print("ERROR: Image failed to load. Please check the file name and path.")
    print("Ensure 'input_image.jpg' is in the same folder as this script.")
    print("---------------------------------------------------------")
else:
    print("Image loaded successfully. Continuing processing...")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 3. Resize the image (width=150, height=100)
resized_img = cv2.resize(gray_img, (150, 100), interpolation=cv2.INTER_LINEAR)

cv2.imshow('Original Image', img)
cv2.imshow('Resized Grayscale Image', resized_img)
```

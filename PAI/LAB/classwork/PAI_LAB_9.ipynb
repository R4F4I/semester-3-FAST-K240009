# linear regression from scratch


```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
np.random.seed(42)
X = np.random.rand(50) * 10
y = 2 * X + 1 + np.random.randn(50) * 5

# Create the first plot: Line not matching the trend
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X, y, label='Data Points')
plt.plot(X, 0.5 * X + 10, color='red', linestyle='--', label='Misleading Line') # Line not matching
plt.title('Scatter Plot with Misleading Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

# Create the second plot: Line matching the trend
plt.subplot(1, 2, 2)
plt.scatter(X, y, label='Data Points')
# Fit a simple linear regression line for the trend
m, b = np.polyfit(X, y, 1)
plt.plot(X, m * X + b, color='green', linestyle='-', label='Trend Line') # Line matching
plt.title('Scatter Plot with Trend Line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## about

- suppose a dataset, $(x_0,y_0),(x_1,y_1),(x_2,y_2),...(x_n,y_n)$, with a linear trend
- linear regression will do its best to get a line to match the trend by nudging only 2 parameters:
  - the gradient of the line
  - the y-intercept
$$
h(x)=mx+b
$$


- the parameters can also be list:
$$
\theta = (m,b)
$$

- the line will match when the sum dist of the line with all the points is minimal
$$
\small\text{minimize} \large \sum^{n}_{i=0} (y_i - h(x_i))^2
$$

- where $h(x_i)$ is the height of line at $x_i$


# implementation

## prerequistites

```python
class pair:
  def __init__(self,x,y):
    self.x = x
    self.y = y

def convert_to_datset(arr1,arr2,size):
  # return a arr of pairs
  # size MUST be same
  dataset = []
  for i in range(size):
    dataset.append(pair(arr1[i],arr2[i]))
  return dataset

class params:
  def __init__(self,m,b):
    self.m = m
    self.b = b

# function of the line to be matched (hypothesis function)
def h(x,params):
  return params.m*x+params.b

```

Data Analysis for the Titanic Dataset with Pandas and Matplotlib
What is Matplotlib?
Its main job is to take your data (like numbers in a list or a Pandas BataFrame) and turn it into a static picture (a plot or chart).
It's incredibly powerful and customizable. Almost any plot you can imagine, you can build with Matplotlib.
The Most Important Concept: fig and ax
fig, ax plt.subplots(). This is the "object-oriented" way to plot, and it's the most flexible..
Let's break down that one line:

```python
fig, ax plt.subplots(figsize=(18,6))
```

plt.subplots(): This is a function that creates two things for you:

1.  Fig (The Figure): This is the entire window or page your plot lives on.
    Think of it as the picture frame. You use fig to do things like save the plot to a file (e.g., fig.savefig('my\_plot.png')).
2.  ax (The Axes): This is the actual plot itself-the part with the data, labels, and ticks.
    Think of it as the canvas inside the frame This is what you '11 use 99% of the time.
    You call methods on ax to create the plot, like
    ax.bar(), ax, scatter(), and ax.set title().
    Analogy: fig is the picture frame, ax is the canvas you draw on.

## Demonstrate matplotlib features (bar plot)

### Subtask:
Create a bar plot using the Titanic dataset to visualize survival rates by passenger class, as suggested in the Matplotlib-Titanic.pdf.


**Reasoning**:
Import the necessary library, calculate the survival rate by passenger class, and create a bar plot to visualize these rates with appropriate labels and a title.



````python
import matplotlib.pyplot as plt

# Calculate survival rate by passenger class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Create the bar plot
fig, ax = plt.subplots(figsize=(8, 6))
survival_rate_by_class.plot(kind='bar', ax=ax, color=['skyblue', 'lightgreen', 'salmon'])

# Add title and labels
ax.set_title('Survival Rate by Passenger Class')
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Survival Rate')

# Rotate x-axis labels for better readability
plt.xticks(rotation=0)

# Display the plot
plt.tight_layout()
plt.show()
````

## Demonstrate matplotlib features (histogram)

### Subtask:
Create a histogram using the Titanic dataset to visualize the distribution of passenger ages.


**Reasoning**:
Generate a histogram of the 'Age' column to visualize its distribution, add appropriate labels and a title, and display the plot.



````python
# Create a figure and axes for the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Generate a histogram of the 'Age' column
# Drop NaN values in 'Age' before plotting
ax.hist(df['Age'].dropna(), bins=30, color='skyblue', edgecolor='black')

# Add title and labels
ax.set_title('Distribution of Passenger Ages')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')

# Display the plot
plt.tight_layout()
plt.show()
````

````python
# This cell demonstrates creating a bar plot using matplotlib to visualize the
# survival rate by passenger class in the Titanic dataset.

# Import the matplotlib.pyplot library for plotting.
import matplotlib.pyplot as plt

# Calculate the survival rate for each passenger class by grouping the DataFrame
# by 'Pclass' and taking the mean of the 'Survived' column.
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Create a figure and axes object for the plot. The figsize argument controls the size of the plot.
fig, ax = plt.subplots(figsize=(8, 6))

# Generate a bar plot using the calculated survival rates.
# 'kind='bar'' specifies a bar plot. 'ax=ax' draws the plot on the created axes.
# 'color' sets the colors for the bars.
survival_rate_by_class.plot(kind='bar', ax=ax, color=['skyblue', 'lightgreen', 'salmon'])

# Set the title of the plot.
ax.set_title('Survival Rate by Passenger Class')
# Set the label for the x-axis (Passenger Class).
ax.set_xlabel('Passenger Class')
# Set the label for the y-axis (Survival Rate).
ax.set_ylabel('Survival Rate')

# Rotate the x-axis labels to prevent overlap, making them easier to read.
plt.xticks(rotation=0)

# Adjust layout to prevent labels and title from overlapping.
plt.tight_layout()
# Display the generated plot.
plt.show()
````
